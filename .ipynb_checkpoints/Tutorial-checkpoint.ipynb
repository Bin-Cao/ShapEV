```python
from ShapEV import EVsearch
```

```python
# Instantiate ShapEV
model = EVsearch.EVkit('T91-LBE-Train.csv',)
```

```python
"""
feature_set (list, optional): List of feature names to focus on. If not provided, it uses 
                               all available features.
"""

# The equivalent value will be constructed using the assigned feature set. 
# If you want the equivalent value to exclude specific features, please specify a feature_set without them.

set = ['Si','CT','Ct ']

model = EVsearch.EVkit('T91-LBE-Train.csv',feature_set = set)
```

```python
# training the model and calculate the joint shap 
model_type  = 'XGBoost'
model.fit( model_type,n_iter=500)
```

```python
# cal the Equivalent value:
model.shap(SPACE=7)
```
